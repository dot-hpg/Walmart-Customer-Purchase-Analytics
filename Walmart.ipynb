# 6. Purchase Distribution, Central Tendency and Outlier Analysis
### Objective: 
To understand the distribution of customer purchase amounts, identify outliers, and analyze key summary statistics such as mean, median, quartiles, and variability.

```python
df['Purchase'].describe()
```

```python
q1 = df['Purchase'].quantile(0.25)
q3 = df['Purchase'].quantile(0.75)

iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = df[
    (df['Purchase'] < lower_bound) |
    (df['Purchase'] > upper_bound)
]

print("Number of Outliers:", len(outliers))

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Statistics
min_val = df['Purchase'].min()
q1 = df['Purchase'].quantile(0.25)
median = df['Purchase'].median()
q3 = df['Purchase'].quantile(0.75)
max_val = df['Purchase'].max()

mean = df['Purchase'].mean()
iqr = q3 - q1

# Whiskers
lower_fence = q1 - 1.5 * iqr
upper_fence = q3 + 1.5 * iqr

lower_whisker = df.loc[df['Purchase'] >= lower_fence, 'Purchase'].min()
upper_whisker = df.loc[df['Purchase'] <= upper_fence, 'Purchase'].max()

# Plot
plt.figure(figsize=(18, 6))

sns.boxplot(
    x='Purchase',
    data=df,
    width=0.6,
    showmeans=True
)

# Mean Line
plt.axvline(
    mean,
    color='red',
    linestyle='--',
    linewidth=2,
    label=f'Mean = ₹{mean:.0f}'
)

# Labels above the boxplot
label_y = 0.22

stats = [
    (min_val, f'Min\n{min_val:.0f}'),
    (lower_whisker, f'LW\n{lower_whisker:.0f}'),
    (q1, f'Q1\n{q1:.0f}'),
    (median, f'Median\n{median:.0f}'),
    (q3, f'Q3\n{q3:.0f}'),
    (upper_whisker, f'UW\n{upper_whisker:.0f}'),
    (max_val, f'Max\n{max_val:.0f}')
]

for x, label in stats:
    plt.text(
        x,
        label_y,
        label,
        ha='center',
        va='bottom',
        fontsize=11,
        fontweight='bold'
    )

plt.title(
    'Purchase Distribution with Five-Number Summary and Whiskers',
    fontsize=18,
    fontweight='bold',
    pad=20
)

plt.xlabel('Purchase Amount (₹)', fontsize=14)
plt.yticks([])
plt.legend(fontsize=12)

plt.tight_layout()
plt.show()

print(f"IQR: {iqr:.2f}")
```

### Observation, Insight & Recommendation:
The purchase distribution is positively skewed, with a mean purchase amount of ₹9,264 exceeding the median of ₹8,047, indicating that a small number of high-spending customers increase the average purchase value. Most purchases fall between ₹5,823 (Q1) and ₹12,054 (Q3), while the presence of outliers beyond the upper whisker confirms unusually large transactions. # 6. Statistical Analysis Using Central Limit Theorem (CLT)

### 6.1 Objective
The objective of this analysis is to understand how the sampling distribution of average purchase amounts behaves for male customers. Using the Central Limit Theorem (CLT), we will repeatedly draw random samples and observe whether the distribution of sample means approaches a normal distribution as sample size increases.


### 6.2 Population Statistics – Male Customers

```python
male_purchase = df[df['Gender'] == 'M']['Purchase']

print("Population Count :", male_purchase.count())
print("Population Mean  :", male_purchase.mean())
print("Population Std   :", male_purchase.std())
```

### Observation:
The male customer population consists of 414,259 transactions with an average purchase amount of ₹9,437.53 and a standard deviation of ₹5,092.19. These statistics serve as the benchmark for evaluating the sampling distributions generated in subsequent analyses.


### 6.3 Sampling Distribution of Mean Purchases (n = 30)

```python
sample_means_male_30 = []

for i in range(1000):
    sample = male_purchase.sample(
        n=30,
        replace=True
    )
    
    sample_means_male_30.append(
        sample.mean()
    )
```

```python
len(sample_means_male_30)
```

```python
# Visualization:

plt.figure(figsize=(8,5))

sns.histplot(
    sample_means_male_30,
    kde=True
)

plt.title(
    "Sampling Distribution of Male Mean Purchases (n=30)"
)

plt.xlabel("Sample Mean")
plt.ylabel("Frequency")

plt.show()
```

```python
print("Sampling Mean :", np.mean(sample_means_male_30))
print("Sampling Std  :", np.std(sample_means_male_30))
```

### Observation, Insight & Recommendation:
The sampling distribution of mean purchase amounts for male customers appears approximately normal despite the original purchase distribution being right-skewed. The distribution is centered around the population mean (~₹9,438), demonstrating the Central Limit Theorem. The spread of the sampling distribution is significantly smaller than the original purchase distribution, indicating that sample means are more stable than individual transactions.

# 7. Confidence Interval Analysis

### 7.1 Objective
The objective of this section is to estimate the range within which the true average purchase amount of male customers is likely to lie. Using a 95% confidence interval, we can quantify the uncertainty associated with the sample mean and make population-level inferences.


### 7.2 Confidence Interval for Male Customers

```python
# Calculate Population Statistics

mean_m = male_purchase.mean()
std_m = male_purchase.std()
n_m = len(male_purchase)

print("Mean:", mean_m)
print("Standard Deviation:", std_m)
print("Sample Size:", n_m)
```

```python
# Calculate Standard Error

se_m = std_m / np.sqrt(n_m)

print("Standard Error:", se_m)
```

```python
# Calculate 95% Confidence Interval

import scipy.stats as stats

ci_95_m = stats.norm.interval(
    confidence=0.95,
    loc=mean_m,
    scale=se_m
)

print("95% Confidence Interval:", ci_95_m)
```

### Observation, Insight & Recommendation:
The average purchase amount for male customers is ₹9,437.53. Using a 95% confidence interval, the true population mean purchase amount is estimated to lie between ₹9,422.02 and ₹9,453.03.

The confidence interval is relatively narrow, indicating high precision in estimating the population mean. This is primarily due to the large sample size of over 414,000 transactions. Therefore, Walmart can be highly confident that the true average spending of male customers lies within this interval.


### 7.3 Visualization of Confidence Interval

### Objective:

Visualize the 95% Confidence Interval for Male Customers and understand the range within which the true population mean is expected to lie.

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(
    mean_m - 4*se_m,
    mean_m + 4*se_m,
    1000
)

y = norm.pdf(x, mean_m, se_m)

plt.figure(figsize=(10,5))

plt.plot(x, y, linewidth=2)

plt.fill_between(
    x,
    y,
    where=(x >= ci_95_m[0]) & (x <= ci_95_m[1]),
    alpha=0.3,
    label='95% Confidence Interval'
)

plt.axvline(
    mean_m,
    linestyle='--',
    linewidth=2,
    label=f'Mean = ₹{mean_m:.2f}'
)

plt.axvline(
    ci_95_m[0],
    linestyle=':',
    linewidth=2,
    label=f'Lower CI = ₹{ci_95_m[0]:.2f}'
)

plt.axvline(
    ci_95_m[1],
    linestyle=':',
    linewidth=2,
    label=f'Upper CI = ₹{ci_95_m[1]:.2f}'
)

plt.title('95% Confidence Interval for Male Customer Purchase Amount')
plt.xlabel('Average Purchase Amount (₹)')
plt.ylabel('Density')
plt.legend()

plt.show()
```

### Observation, Insight & Recommendation:
The normal distribution curve is centered around the average purchase amount of male customers (₹9,437.53). The shaded region represents the 95% confidence interval, indicating that the true population mean purchase amount is expected to lie between ₹9,422.02 and ₹9,453.03. The narrow interval reflects high precision due to the large sample size.

### 7.4 Key Insights

1. The average purchase amount of male customers is ₹9,437.53 per transaction.

2. The 95% confidence interval for male customer spending ranges from ₹9,422.02 to ₹9,453.03.

3. The confidence interval is very narrow, indicating a highly precise estimate of the population mean.

4. The large sample size (414,259 transactions) significantly reduces sampling error and increases confidence in the estimated spending behavior.

5. Walmart can reliably use this estimate to understand and predict the spending patterns of male customers during Black Friday sales.

```python
ci_mar = stats.norm.interval(
    confidence=0.95,
    loc=mean_mar,
    scale=se_mar
)

ci_unmar = stats.norm.interval(
    confidence=0.95,
    loc=mean_unmar,
    scale=se_unmar
)

print("Married CI :", ci_mar)
print("Unmarried CI :", ci_unmar)
```

### Observation:
The average purchase amount for married customers is ₹9,261.17, while the average purchase amount for unmarried customers is ₹9,265.91. The 95% confidence interval for married customers ranges from ₹9,240.46 to ₹9,281.89, whereas the confidence interval for unmarried customers ranges from ₹9,248.62 to ₹9,283.20. The confidence intervals overlap substantially, suggesting that marital status has little impact on customer spending behavior during Walmart's Black Friday sales event.

### 11.4 Standard Error Calculation

```python
# Calculate Standard Error for Each Age Group

age_ci_data = []

for age in df['Age'].unique():

    age_data = df[df['Age']==age]['Purchase']

    mean = age_data.mean()
    std = age_data.std()
    n = len(age_data)

    se = std / np.sqrt(n)

    age_ci_data.append([
        age,
        mean,
        std,
        n,
        se
    ])

age_ci_df = pd.DataFrame(
    age_ci_data,
    columns=[
        'Age',
        'Mean',
        'Std',
        'Count',
        'SE'
    ]
)

age_ci_df
```

### 11.5 Confidence Interval Analysis

```python
# Calculate 95% Confidence Intervals
age_ci_results = []

for age in df['Age'].unique():

    age_data = df[df['Age']==age]['Purchase']

    mean = age_data.mean()
    std = age_data.std()
    n = len(age_data)

    se = std / np.sqrt(n)

    ci = stats.norm.interval(
        confidence=0.95,
        loc=mean,
        scale=se
    )

    age_ci_results.append([
        age,
        mean,
        ci[0],
        ci[1]
    ])

age_ci_results = pd.DataFrame(
    age_ci_results,
    columns=[
        'Age',
        'Mean',
        'Lower_CI',
        'Upper_CI'
    ]
)

age_ci_results
```

### 11.6 Confidence Interval Visualization

```python
# Compare Age Groups
plt.figure(figsize=(12,6))

for i, row in age_ci_results.iterrows():

    plt.errorbar(
        row['Mean'],
        i,
        xerr=[[row['Mean']-row['Lower_CI']],
              [row['Upper_CI']-row['Mean']]],
        fmt='o',
        capsize=5
    )

plt.yticks(
    range(len(age_ci_results)),
    age_ci_results['Age']
)

plt.xlabel('Average Purchase Amount (₹)')
plt.title('95% Confidence Intervals by Age Group')

plt.grid(alpha=0.3)

plt.show()
```
